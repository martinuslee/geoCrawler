```python
from Bio import Entrez
from selenium import webdriver
import os
import sys
# 라이브러리 임포팅

#handle = Entrez.einfo()
handle = Entrez.einfo(db="gds") # DB 선택

Entrez.email = 'supreme034@korea.ac.kr'

record = Entrez.read(handle)
record.keys()


desc = record['DbInfo']['Description']
cnt = record['DbInfo']['Count']
upd = record['DbInfo']['LastUpdate']

print(desc + '\n' +cnt + '\n' + upd+ '\n')

#데이터셋 정보 출력 
```

```python
Entrez.email = "supreme034@korea.ac.kr" # 꼭 본인 이메일을 써줘야함
handle = Entrez.esearch(db="gds", term="circadian rhythm") # 디비 와 원하는 검색어 
record = Entrez.read(handle)
gdsId = []
for i in record["IdList"]:
    gdsId.append(i)
    
gdsId
# 검색 결과의 고유 아이디 출력 
```

```python
for i in gdsId:
    handle=(Entrez.efetch(db="gds", id=i, rettype="gb", retmode="text"))
    print(handle.read())
    
# 대략적인 요약들
```
